```python
import pandas as pd
import numpy as np

file_path = '..\\data\\sales-data-raw.csv'
df = pd.read_csv(file_path, index_col='ADDRESS') 

df = df[~df.index.duplicated(keep='first')]

df.head()
```

```python
#capitalize addresses (i.e. index column) and remove trailing space 
df.index = df.index.str.upper()
df.index = df.index.str.strip()
```

```python
#clean LIST_PRICE and SELL_PRICE columns
df['LIST_PRICE'] = df['LIST_PRICE'].str.replace('$', '', regex=True)
df['LIST_PRICE'] = df['LIST_PRICE'].str.replace(',', '', regex=True)
df['LIST_PRICE'] = df['LIST_PRICE'].replace('-', np.nan, regex=True)
df['LIST_PRICE'] = df['LIST_PRICE'].astype(float)

df['SELL_PRICE'] = df['SELL_PRICE'].str.replace('$', '', regex=True)
df['SELL_PRICE'] = df['SELL_PRICE'].str.replace(',', '', regex=True)
df['SELL_PRICE'] = df['SELL_PRICE'].replace('-', np.nan, regex=True)
df['SELL_PRICE'] = df['SELL_PRICE'].astype(float)

#clean SOLD_ON column
df['SOLD_ON'] = df['SOLD_ON'].str.replace("Sold on ", '', regex=True)
df['SOLD_ON'] = df['SOLD_ON'].str.replace("[", '', regex=True)
df['SOLD_ON'] = df['SOLD_ON'].str.replace("'", '', regex=True)
df['SOLD_ON'] = df['SOLD_ON'].str.replace("]", '', regex=True)
df['SOLD_ON'] = df['SOLD_ON'].str.replace(" 2021", ', 2021', regex=True)

df['SOLD_ON'] = df['SOLD_ON'].str.replace("Jan", 'January', regex=True)
df['SOLD_ON'] = df['SOLD_ON'].str.replace("Feb", 'February', regex=True)
df['SOLD_ON'] = df['SOLD_ON'].str.replace("Mar", 'March', regex=True)
df['SOLD_ON'] = df['SOLD_ON'].str.replace("Apr", 'April', regex=True)
df['SOLD_ON'] = df['SOLD_ON'].str.replace("Jun", 'June', regex=True)
df['SOLD_ON'] = df['SOLD_ON'].str.replace("Jul", 'July', regex=True)
df['SOLD_ON'] = df['SOLD_ON'].str.replace("Aug", 'August', regex=True)
df['SOLD_ON'] = df['SOLD_ON'].str.replace("Sep", 'September', regex=True)
df['SOLD_ON'] = df['SOLD_ON'].str.replace("Oct", 'October', regex=True)
df['SOLD_ON'] = df['SOLD_ON'].str.replace("Nov", 'November', regex=True)
df['SOLD_ON'] = df['SOLD_ON'].str.replace("Dec", 'December', regex=True)
```

```python
#clean DOM column
df['DOM'] = df['DOM'].replace('-', np.nan, regex=True)
df['DOM'] = df['DOM'].astype(float)

#clean BDRMS column
df['BDRMS'] = df['BDRMS'].replace('-', np.nan, regex=True)
df['BDRMS'] = df['BDRMS'].astype(float)

#clean BDRMS+ column
df['BDRMS+'] = df['BDRMS+'].replace('-', np.nan, regex=True)
df['BDRMS+'] = df['BDRMS+'].astype(float)
```

```python
#clean SIZE_SQFT column
df['SIZE_SQFT'] = df['SIZE_SQFT'].str.replace(r'(?<=-)[0-9]', '', regex=True)
df['SIZE_SQFT'] = df['SIZE_SQFT'].str.replace(r'(?<=-)[0-9]', '', regex=True)
df['SIZE_SQFT'] = df['SIZE_SQFT'].str.replace(r'(?<=-)[0-9]', '', regex=True)
df['SIZE_SQFT'] = df['SIZE_SQFT'].str.replace(r'(?<=-)[0-9]', '', regex=True)
df['SIZE_SQFT'] = df['SIZE_SQFT'].str.replace('< ', '', regex=True)
df['SIZE_SQFT'] = df['SIZE_SQFT'].str.replace('-', '', regex=True)

df['SIZE_SQFT'] = df['SIZE_SQFT'].replace('', np.nan, regex=True)
df['SIZE_SQFT'] = df['SIZE_SQFT'].astype(float)
```

```python
#clean AGE column
df['AGE'] = df['AGE'].str.replace(' years', '', regex=True)
df['AGE'] = df['AGE'].str.replace(r'[0-9](?=-)', '', regex=True)
df['AGE'] = df['AGE'].str.replace(r'[0-9](?=-)', '', regex=True)
df['AGE'] = df['AGE'].str.replace('-', '', regex=True)
df['AGE'] = df['AGE'].str.replace('+', '', regex=True)
df['AGE'] = df['AGE'].str.replace('New', '0', regex=True)

df['AGE'] = df['AGE'].replace('', np.nan, regex=True)
df['AGE'] = df['AGE'].astype(float)
```

```python
#clean ASS_YR column
df['ASS_YR'] = df['ASS_YR'].replace('-', '', regex=True)
df['ASS_YR'] = df['ASS_YR'].replace('', np.nan, regex=True)
df['ASS_YR'] = df['ASS_YR'].astype(float)

#clean ASS_AMT column
df['ASS_AMT'] = df['ASS_AMT'].replace('-', '', regex=True)
df['ASS_AMT'] = df['ASS_AMT'].replace('', np.nan, regex=True)
df['ASS_AMT'] = df['ASS_AMT'].astype(float)

#clean TAX_YR column
df['TAX_YR'] = df['TAX_YR'].replace('-', '', regex=True)
df['TAX_YR'] = df['TAX_YR'].replace('', np.nan, regex=True)
df['TAX_YR'] = df['TAX_YR'].astype(float)

#clean TAX_AMT column
df['TAX_AMT'] = df['TAX_AMT'].replace('-', '', regex=True)
df['TAX_AMT'] = df['TAX_AMT'].replace('', np.nan, regex=True)
df['TAX_AMT'] = df['TAX_AMT'].astype(float)
```

```python
#clean LOT_DPTH column
df['LOT_DPTH'] = df['LOT_DPTH'].replace('-', '', regex=True)
df['LOT_DPTH'] = df['LOT_DPTH'].replace('', np.nan, regex=True)
df['LOT_DPTH'] = df['LOT_DPTH'].astype(float)

#clean LOT_FRNT column
df['LOT_FRNT'] = df['LOT_FRNT'].replace('-', '', regex=True)
df['LOT_FRNT'] = df['LOT_FRNT'].replace('', np.nan, regex=True)
df['LOT_FRNT'] = df['LOT_FRNT'].astype(float)
```

```python
#convert LON and LAT to float
df['LON'] = df['LON'].apply(lambda x : np.nan if x == '-' else x)
df['LON'] = df['LON'].astype(float)

df['LAT'] = df['LAT'].apply(lambda x : np.nan if x == '-' else x)
df['LAT'] = df['LAT'].astype(float)
```

```python
import csv
file_path = '..\\data\\raw\\addresses-base.csv'

with open(file_path, encoding='utf-8', newline='') as file:
    addresses_base = list(csv.reader(file))

addresses_base[:10]
```

```python
def nested_list_contains(nl, target):
    for thing in nl:
        if type(thing) is list:
            if nested_list_contains(thing, target):
                return thing
        if thing == target:
            return thing
    return False
```

```python
index = df.index
for i in index:
    try:
        address_to_find = i.strip()
        thing = nested_list_contains(addresses_base, address_to_find)
        add_id = thing[0]
        df.at[i, 'ADD_ID'] = add_id
    except:
        df.at[i, 'ADD_ID'] = np.nan

df.head()
```

```python
import requests

def get_coords(address_to_fetch):

    url = 'https://nominatim.openstreetmap.org/search/' + address_to_fetch +'?format=json'

    response = requests.get(url).json()
    lat = response[0]["lat"]
    lon = response[0]["lon"]

    return lat, lon

index = df.index
for address in index:
    try:
        address_to_fetch = address + "Guelph Ontario"
        lat, lon = get_coords(address_to_fetch)
        df.at[address, 'LAT'] = lat
        df.at[address, 'LON'] = lon
    except:
        df.at[address, 'LAT'] = np.nan
        df.at[address, 'LON'] = np.nan
```
